```python
import os
import pandas as pd
```

```python
processed_path = 'drive/MyDrive/RSS/Processed/'
folders = os.listdir(processed_path)
```

```python
def get_tex_files(source):
    matches = []
    for root, dirnames, filenames in os.walk(source):
        for filename in filenames:
            if filename.endswith(('.tex')):
                matches.append(os.path.join(root, filename))
    return matches
```

```python
data = []
startflag = '\\begin{equation}'
endflag = '\\end{equation}'

data2 = []
startflag2 = '\\begin{algorithm}'
endflag2 = '\\end{algorithm}'

filenum = 3000

for folder in folders[3000:6000]:
    print('Processing File', filenum)
    filenum += 1
    folder_path = processed_path + folder + '/'
    files = get_tex_files(folder_path)
    for fp in files:
        try:
            content = open(fp, 'r').read()
            content2 = content
            ind = content.find(startflag)
            while not ind == -1:
                content = content[ind+len(startflag):]
                end = content.find(endflag)
                eqn = content[:end]
                datapt = [folder, eqn]
                data.append(datapt)
                ind = content.find(startflag)
            
            content = content2
            ind = content.find(startflag2)
            while not ind == -1:
                content = content[ind+len(startflag2):]
                end = content.find(endflag2)
                alg = content[:end]
                datapt = [folder, alg]
                data2.append(datapt)
                ind = content.find(startflag2)
        except:
            print('Failed on file: ', fp)
```

```python
df = pd.DataFrame(data, columns=['id', 'eqn'])
df.to_csv('equations2.csv')
df.head()
```

```python
df = pd.DataFrame(data2, columns=['id', 'alg'])
df.to_csv('algorithms2.csv')
df.head()
```
