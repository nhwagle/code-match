```python
import os
import io

import requests
from PyPDF2 import PdfFileReader
import pandas as pd
```

```python
datalinkspath = './paper-code-links.csv'

datalinksdf = pd.read_csv(datalinkspath)
datalinksdf.head()
```

```python
datalinksdf.shape
```

```python
contents = []
sample = 0
for url in datalinksdf['link'][:100]:
    sample += 1
    print(sample)
    try:
        r = requests.get(url)
        f = io.BytesIO(r.content)
        reader = PdfFileReader(f)
        content = ''
        numpgs = reader.getNumPages()
        for i in range(numpgs):
            page = reader.getPage(i).extractText()
            content += page
        contents.append(content)
    except:
        print(url) 
        contents.append('Text Extraction Failed')
```

```python
extractionsdf = datalinksdf[:100]
extractionsdf['text'] = contents
extractionsdf.head()
```

```python
extractionsdf.to_csv('text-extractions-100.csv')
```
